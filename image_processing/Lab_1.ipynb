```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import numpy as np
from skimage import img_as_float
from skimage.io import imread
from skimage.color import rgb2lab, lab2rgb
from skimage.util import invert
from skimage.exposure import equalize_hist
import matplotlib.pylab as plt
```

```python
img = imread("/content/drive/MyDrive/lab-1/pexels-pixabay-56866.jpg")
plt.imshow(img)
plt.title("image")
plt.axis("off")
plt.show()
```

```python
R= img[:, :, 0]
G= img[:, :, 1]
B= img[:, :, 2]

output = [img, R, G, B]
titles = ['Original image', 'R channel', 'G channel', 'B channel']

for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.axis('off')
    plt.title(titles[i])
    if i == 0:
        plt.imshow(output[i])
    else:
        plt.imshow(output[i], cmap='gray')


plt.show()
```

```python
t4 = imread("/content/drive/MyDrive/lab-1/t4.jpg")
plt.imshow(t4)
plt.title("t4")
plt.axis("off")
plt.show()
```

```python
t5 = imread("/content/drive/MyDrive/lab-1/t5.jpg")
plt.imshow(t5)
plt.title("t5")
plt.axis("off")
plt.show()
```

```python
from skimage.transform import resize
# Resize t4 to match the dimensions of t5
t4_resized = resize(t4, (t5.shape[0], t5.shape[1]), anti_aliasing=True) # Resize t4 to the shape of t5

# Now you can perform the subtraction
diff_image = t5 - t4_resized

# Display the resulting image
plt.figure()
plt.imshow(diff_image)
plt.title("Motion Difference")
plt.axis("off")
plt.show()
```

```python
from skimage import exposure

# Darker Image
darker_image = exposure.adjust_gamma(img, gamma=2)  # Increase gamma for darker image

# Reduced Contrast
reduced_contrast_image = exposure.equalize_hist(img)  # Histogram equalization

# Inverted Image
inverted_image = invert(img)  # Invert pixel values

# Brighter Image
brighter_image = exposure.adjust_gamma(img, gamma=0.2)  # Decrease gamma for brighter image

# Increased Contrast
increased_contrast_image = exposure.adjust_sigmoid(img) # Adjust sigmoid curve for contrast

# Display the images
plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 3, 2)
plt.imshow(darker_image)
plt.title("Darker Image")
plt.axis("off")

plt.subplot(2, 3, 3)
plt.imshow(reduced_contrast_image)
plt.title("Reduced Contrast")
plt.axis("off")

plt.subplot(2, 3, 4)
plt.imshow(inverted_image)
plt.title("Inverted Image")
plt.axis("off")

plt.subplot(2, 3, 5)
plt.imshow(brighter_image)
plt.title("Brighter Image")
plt.axis("off")

plt.subplot(2, 3, 6)
plt.imshow(increased_contrast_image)
plt.title("Increased Contrast")
plt.axis("off")

plt.tight_layout()
plt.show()
```
