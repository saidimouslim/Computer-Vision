```python
G4 = np.array([[-1,-1,0], [-1,3,0], [0,0,0]])
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
sobel_horizontal = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=5)
```

```python
sobel_vertical = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=5)
```

```python
sobel_vertical = cv2.convertScaleAbs(sobel_vertical)
sobel_horizontal = cv2.convertScaleAbs(sobel_horizontal)
```

```python
plt.title('Sobel Horizontal')
plt.imshow(sobel_horizontal, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.title('Sobel Vertical')
plt.imshow(sobel_vertical, cmap='gray')
plt.axis('off')
plt.show()
```
